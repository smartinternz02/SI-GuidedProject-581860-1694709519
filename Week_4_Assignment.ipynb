# Data Preprocessing

## Import the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing the DataSet

```python
df=pd.read_csv("/content/WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Checking for null values

```python
df.isnull().sum()
```

## Data Visualization

```python
sns.countplot(x=df.Attrition,data=df)
```

```python
corr=df.corr(numeric_only=True)
corr
```

```python
fig=plt.figure(figsize=(10,10))
sns.heatmap(corr,annot=True)
```

## Outlier Detection

```python
fig=plt.figure(figsize=(60,15))
sns.boxplot(df)
```

```python
for i in df.columns:
    if(df[i].dtype!=object):
        plt.figure()
        plt.title(i)
        sns.boxplot(df[i])
```

```python
l=["MonthlyIncome","NumCompaniesWorked","StockOptionLevel","TotalWorkingYears","TrainingTimesLastYear","YearsAtCompany", "YearsInCurrentRole", "YearsSinceLastPromotion",
       "YearsWithCurrManager"]
for i in l:
    q3=df[i].quantile(0.75)
    q1=df[i].quantile(0.25)
    IQR=q3-q1
    upper_limit=q3+(1.5*IQR)
    lower_limit=q1-(1.5*IQR)
    df[i]=np.where(df[i]>upper_limit,df[i].median(),df[i])
```

```python
for i in df.columns:
    if(df[i].dtype!=object):
        plt.figure()
        plt.title(i)
        sns.boxplot(df[i])
```

## Splitting Dependent and Independent variables

```python
df.drop(columns=['EmployeeCount','StandardHours','EmployeeNumber','Over18'],axis=1,inplace=True)
```

```python
df.head(5)
```

```python
x=df.drop("Attrition",axis=1)
y=df.Attrition
```

```python
x.head()
```

```python
y.head()
```

```python
type(x)
```

```python
type(y)
```

## Perform Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y_c=pd.Series(le.fit_transform(y))
```

```python
q=["BusinessTravel","Department","EducationField","Gender","JobRole","MaritalStatus","OverTime"]
for q1 in q:
  x[q1]=x[q1].astype('category').cat.codes
```

```python
x.head()
```

```python
y_c.head()
```

## Feature Scaling.

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_scaled=pd.DataFrame(ss.fit_transform(x),columns=x.columns)
```

```python
x_scaled.head()
```

## Splitting Data into Train and Test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y_c,test_size=0.2,random_state=0)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

# Preprocessing Done

# -->Model Building - Logistic regression

## Import the model building Libraries

```python
from sklearn.linear_model import LogisticRegression
```

## Initializing the model

```python
lr=LogisticRegression()
```

## Training the Model

```python
lr.fit(x_train,y_train)
```

## Testing the Model

```python
y_pred=lr.predict(x_test)
```

```python
pd.DataFrame({"Actual_values":y_test,"Predicted_values":y_pred})
```

## Evaluation of Model & Performance metrics

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve
```

```python
print("Accuracy of model :",accuracy_score(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
pd.crosstab(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
#ROC-AUC Curve
probability=lr.predict_proba(x_test)[:,1]
fpr,tpr,threshsholds = roc_curve(y_test,probability)
plt.plot(fpr,tpr)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC CURVE')
plt.show()
```

```python
threshsholds
```
